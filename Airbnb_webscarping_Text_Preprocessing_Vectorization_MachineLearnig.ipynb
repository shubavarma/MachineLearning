```python
import requests
import pandas as pd
import json
import re

listing_ids = ['20445002', '3834276',"24415592",'48534993','24059136','19670926','43120947','32234028','3545435','4402940','6168679','2126708','3545435']
url_template = 'https://www.airbnb.com/api/v2/reviews?key=d306zoyjsyarp7ifhu67rjxn52tv0t20&currency=INR&locale=en&_format=for_p3&_limit=50&_offset={offset}&listing_id={listing_id}&role=guest&_order=language_country'
df = pd.DataFrame(columns=['Title','Location','reviewees','reviewers','reviews', 'ratings'])

for listing_id in listing_ids:
    template_url2 = f'https://www.airbnb.co.in/rooms/plus/{listing_id}?adults=1&category_tag=Tag%3A8524&children=0&enable_m3_private_room=false&infants=0&pets=0&check_in=2023-05-21&check_out=2023-05-26&federated_search_id=39848060-29fe-46f6-b92c-869d91d943ca&source_impression_id=p3_1682922289_oGR%2BulLXLZkuhPYA'
    response=requests.get(template_url2)
    from bs4 import BeautifulSoup
    soup=BeautifulSoup(response.text,"html.parser")
    # to scrape Title
    title1 = re.findall(r"(.+)\s-.+-\sAirbnb$",soup.find("title").text.strip()) 
    title = title1[0]
    # to scrape Location
    location1 = re.findall(r".+\s-(.+)-\sAirbnb$",soup.find("title").text.strip())
    location = location1[0]
    offset = 0
    while True:
        url = url_template.format(offset=offset, listing_id=listing_id)
        response = requests.get(url)

        if response.ok:
            try:
                data = response.json()['reviews']
            except json.decoder.JSONDecodeError:
                print(f'Error decoding JSON for offset {offset} and listing ID {listing_id}')
                break
            titles = []
            locations = []
            reviews = []
            ratings = []
            reviewees = []
            reviewers = []
            for review in data:
                review_text = review['comments']
                rating = review['rating']
                reviewer = review['reviewer']['host_name']
                reviewee = review['reviewee']['host_name']
                titles.append(title)
                locations.append(location)
                reviews.append(review_text)
                ratings.append(rating)
                #print(len(titles),len(ratings))

                reviewers.append(reviewer)
                reviewees.append(reviewee)

            if not reviews:
                break

            reviews_dict = {'Title': titles,'Location': locations,'reviewees': reviewees ,'reviewers': reviewers,'reviews': reviews, 'ratings': ratings }
            df_temp = pd.DataFrame(reviews_dict)
            
            df = pd.concat([df,df_temp], ignore_index=True)
            offset += 50

        else:
            print(f'Error fetching data for offset {offset} and listing ID {listing_id}')
            break

df
```

# Remove Special Characters

```python
def removeSpecialChars(text):
    return re.sub(r'[^0-1a-zA-Z]',' ',text)
```

```python
df['reviews'] = df['reviews'].apply(removeSpecialChars)
df
```

# Remove stop words

```python
def removeStopwords(text):
    import nltk
    from nltk.corpus import stopwords
    nltk.download('stopwords')
    a = []
    for word in text.split(' '):
        if word not in stopwords.words('english'):
            a.append(word)
    return " ".join(a)
```

```python
df['reviews'] = df['reviews'].apply(removeStopwords)
```

# Remove Empty strings

```python
def clean_dataframe(df):
    # Drop rows with empty strings in the 'reviews' column
    df = df[df['reviews'].str.strip().astype(bool)]

    # Drop columns that have all empty strings
    df = df.loc[:, (df != '').any(axis=0)]

    return df
```

# Perform Lemmitization

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
def lemTokens(text):
    lemTokens = []
    for word in text.split(' '):
        if word != ' ':
            lemTokens.append(lem.lemmatize(word))
    return lemTokens
```

```python
df['lemTokens'] = df['reviews'].apply(lemTokens)
df
```

```python
df['lemSentence'] = df['lemTokens'].str.join(' ')
df
```

```python
df.to_csv('Airbnb_lemdata.csv')
```
