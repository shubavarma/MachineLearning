# Convert the PIL.JpegImagePlugin.JpegImageFile to numpy array

```python
ironmanNp = np.array(ironmanPil)
type(ironmanNp)
```

```python
# rotate, crop, flip, resize
# We can see how to use on numpy array and PIL object
```

```python
# rotate on PIL object, needs to angle to be rotated to rotate method
ironmanRotated = ironmanPil.rotate(60)
plt.imshow(ironmanRotated) # here it is getting cropped
```

```python
# not to crop we need to say expand = True
ironmanRotated = ironmanPil.rotate(60, expand = True)
plt.imshow(ironmanRotated)
```

```python
# We can change the background color using fillcolor = () we need to pass tuple of RGB code
ironmanRotated = ironmanPil.rotate(60, expand = True, fillcolor = (255,77,0))
plt.imshow(ironmanRotated)
```

```python
# In Numpy we can rotate only 90 degrees anti-clockwise
# need to use numpy array
plt.imshow(np.rot90(ironmanNp))
```

```python
# if we want to rotate 180 degrees
plt.imshow(np.rot90(ironmanNp,2))
```

```python
# flip ....front to back, top to bottom, both front to back and top to bottom (this is transpose)
ironmanFlippedLR = ironmanPil.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(ironmanFlippedLR)
```

```python
ironmanFlippedTB = ironmanPil.transpose(Image.FLIP_TOP_BOTTOM)
ironmanTranspose = ironmanPil.transpose(Image.TRANSPOSE)

plt.figure(figsize = (15,15))
plt.subplot(1, 3, 1)
plt.imshow(ironmanFlippedLR)
plt.title("Left Right Flipped Ironman")
plt.subplot(1, 3, 2)
plt.imshow(ironmanFlippedTB)
plt.title("Top Bottom Flipped ironman")
plt.subplot(1, 3, 3)
plt.imshow(ironmanTranspose)
plt.title("Transposed ironman")
```

```python
# Flip using Numpy array
plt.figure(figsize = (15,15))
plt.subplot(1, 3, 1)
plt.imshow(ironmanNp)
plt.title("Original Ironman")
plt.subplot(1, 3, 2)
plt.imshow(np.fliplr(ironmanNp))
plt.title("Left Right Flipped Ironman")
plt.subplot(1, 3, 3)
plt.imshow(np.flipud(ironmanNp))
plt.title("Top Bottom Flipped ironman")
```

```python
# Adding text to image
from PIL import ImageDraw, ImageFont
```

```python
# https://fonts.google.com/
# use these fonts
# download and unzip
# to add this font we use ImageFont
# we need to mention the font size looking at the scale of the image
font = ImageFont.truetype(r"C:\Users\Kishore\197\MachineLearning\Fira_Sans\FiraSans-Regular.ttf",
                          100)
```

```python
# To draw on the image we need to create ImageDraw object
ironmanDraw = ImageDraw.Draw(ironmanPil)
```

```python
# Adding text to the image
# we need to pass y,x cordinates where we want the text to start
# followed by the text we want to add
# what font to use
# this will be saved on to the Pil object
ironmanDraw.text((100,200),
                text = 'Iron Man',
                font = font )
```

```python
plt.imshow(ironmanPil)
```

```python
# If we want to change the color
ironmanDraw.text((100,200),
                text = 'Iron Man',
                font = font,
                fill = (255,0,0))
plt.imshow(ironmanPil)
```
