```python
# Adding text to image we will import ImageDraw and ImageFont libraries
from PIL import ImageDraw, ImageFont
```

```python
# to add this font we use ImageFont
# we need to mention the font size looking at the scale of the image
font = ImageFont.truetype(r"C:\Users\Kishore\197\MachineLearning\files_fonts\DynaPuff\static\DynaPuff\DynaPuff-Regular.ttf",
                          35)
```

```python
# To draw on the image we need to create ImageDraw object
batmanDraw = ImageDraw.Draw(batmanPil)
```

```python
# Adding text to the image
# we need to pass y,x cordinates where we want the text to start
# followed by the text we want to add
# what font to use
# this will be saved on to the Pil object
# To add Robin's dialogue
batmanDraw.text((10,10),
                text = 'You are too old \n   to be a \n DataScientist',
                font = font,
                fill = (0,0,0))
```

```python
plt.imshow(batmanPil)
```

```python
# To add Batman dialogue
batmanDraw.text((450,10),
                text = 'Learning should \n Never stop... ',
                font = font,
                fill = (0,0,0))
```
